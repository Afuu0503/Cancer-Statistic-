```python
cancer = df
cancer.head()
```

```python
# średnia wszystkich danych 
print("Średni wiek:", df['Age'].mean())
print("Średnia jakość powietrza:", df['Air Pollution'].mean())
print("Średnie zużycie alkoholu:", df['Alcohol use'].mean())
print("Średnia uczulenia na kurz:", df['Dust Allergy'].mean())
```

    Średnia uczulenia na kurz: 5.165

```python
# mediana wszystkich danych 
print("Mediana wiek:", df['Age'].median())
print("Mediana jakość powietrza:", df['Air Pollution'].median())
print("Mediana zużycie alkoholu:", df['Alcohol use'].median())
print("Mediana uczulenia na kurz:", df['Dust Allergy'].median())
```

```python
# Wariancja wszystkich danych 
print("Wariancja wiek:", df['Age'].var())
print("Wariancja jakość powietrza:", df['Air Pollution'].var())
print("Wariancja zużycie alkoholu:", df['Alcohol use'].var())
print("Wariancja uczulenia na kurz:", df['Dust Allergy'].var())
```
