```python
#importowanie bibliotek 
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import math as math
import csv

            
```

```python
# w tym miejscu chciaam okreli konkretne elementy csv i wypisa je w programie
with open("cancer.csv", "r") as csv_file:

        for line in csv.DictReader(csv_file, delimiter=","):
            print("Age:"+ line["Age"] + "Air Pollution:" + line["Air Pollution"]+ "Alcohol use:" + line["Alcohol use"] + "Dust Allergy:" + line["Dust Allergy"] + "Genetic Risk:" + line["Genetic Risk"] + "Level:" + line["Level"] )
```

```python
#w tym miejscu chciaam lepiej zwizualizowa sobie dane za pomoc pandas i dataFrame
df = pd.read_csv("cancer.csv")
```

```python
cancer = df
cancer.head()
```

```python
# rednia wszystkich danych 
print("redni wiek:", df['Age'].mean())
print("rednia jako powietrza:", df['Air Pollution'].mean())
print("rednie zuycie alkoholu:", df['Alcohol use'].mean())
print("rednia uczulenia na kurz:", df['Dust Allergy'].mean())
```

```python
# mediana wszystkich danych 
print("Mediana wiek:", df['Age'].median())
print("Mediana jako powietrza:", df['Air Pollution'].median())
print("Mediana zuycie alkoholu:", df['Alcohol use'].median())
print("Mediana uczulenia na kurz:", df['Dust Allergy'].median())
```

```python
# Wariancja wszystkich danych 
print("Wariancja wiek:", df['Age'].var())
print("Wariancja jako powietrza:", df['Air Pollution'].var())
print("Wariancja zuycie alkoholu:", df['Alcohol use'].var())
print("Wariancja uczulenia na kurz:", df['Dust Allergy'].var())
```
